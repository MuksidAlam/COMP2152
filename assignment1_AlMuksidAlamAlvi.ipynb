```python
"""
Author: Al Muksid Alam Alvi
Assignment: #1
"""

# Initialize gym member variables with appropriate data types
gym_member = "Alex Alliton"              # str - stores member's full name
preferred_weight_kg = 20.5               # float - preferred dumbbell weight in kilograms
highest_reps = 25                        # int - maximum repetitions achieved
membership_active = True                 # bool - current membership status
```

```python

# Display initialized variables
print("=" * 50)
print("GYM MEMBER INFORMATION")
print("=" * 50)
print(f"Member Name: {gym_member}")
print(f"Preferred Weight: {preferred_weight_kg} kg")
print(f"Personal Record (Reps): {highest_reps}")
print(f"Membership Status: {'Active' if membership_active else 'Inactive'}")
print("=" * 50)
```

```python
# dict - Dictionary mapping friend names to workout tuples (yoga, running, weightlifting minutes)
workout_stats = {
    "Alex": (30, 45, 20),      # Alex's workout: 30 min yoga, 45 min running, 20 min weightlifting
    "Jamie": (25, 50, 30),     # Jamie's workout: 25 min yoga, 50 min running, 30 min weightlifting
    "Taylor": (40, 35, 25)     # Taylor's workout: 40 min yoga, 35 min running, 25 min weightlifting
}

# Display initial workout statistics
print("\n" + "=" * 50)
print("INITIAL WORKOUT STATISTICS")
print("=" * 50)
for friend_name, workout_minutes in workout_stats.items():
    print(f"{friend_name:10} -> Yoga: {workout_minutes[0]:2d} min | Running: {workout_minutes[1]:2d} min | Weights: {workout_minutes[2]:2d} min")
print("=" * 50)
```

```python
# Calculate total workout minutes for each friend and add to dictionary
# Using list() to avoid RuntimeError when modifying dictionary during iteration
for friend_name, workout_minutes in list(workout_stats.items()):
    # Only process entries that are tuples (original workout data)
    if isinstance(workout_minutes, tuple):
        total_minutes = sum(workout_minutes)
        # Add new key with "_Total" suffix
        workout_stats[f"{friend_name}_Total"] = total_minutes

# Display updated dictionary with totals
print("\n" + "=" * 50)
print("WORKOUT STATISTICS WITH TOTALS")
print("=" * 50)
for key, value in workout_stats.items():
    if isinstance(value, tuple):
        print(f"{key:15} -> {value}")
    else:
        print(f"{key:15} -> {value} minutes")
print("=" * 50)
```

```python
# list (2D/nested) - Extract workout tuples and convert to list format
# Each row represents a friend's workout data [yoga, running, weightlifting]
workout_list = [
    list(workout_minutes) 
    for workout_minutes in workout_stats.values() 
    if isinstance(workout_minutes, tuple)
]

# Display the 2D list structure
print("\n" + "=" * 50)
print("2D WORKOUT LIST (Row = Friend, Col = Activity)")
print("=" * 50)
activity_headers = ["Yoga", "Running", "Weightlifting"]
print(f"{'Index':<8} {activity_headers[0]:^12} {activity_headers[1]:^12} {activity_headers[2]:^15}")
print("-" * 50)
for index, workout_row in enumerate(workout_list):
    print(f"Friend {index:<2} {workout_row[0]:^12} {workout_row[1]:^12} {workout_row[2]:^15}")
print("=" * 50)
```

```python
# Slice 1: Extract yoga (column 0) and running (column 1) minutes for all friends
yoga_and_running = [workout_row[:2] for workout_row in workout_list]

print("\n" + "=" * 50)
print("SLICE 1: Yoga and Running Minutes (All Friends)")
print("=" * 50)
for index, minutes in enumerate(yoga_and_running):
    print(f"Friend {index}: Yoga = {minutes[0]} min, Running = {minutes[1]} min")
print("=" * 50)

# Slice 2: Extract weightlifting (column 2) minutes for the last two friends
weightlifting_last_two = [workout_row[2] for workout_row in workout_list[-2:]]

print("\n" + "=" * 50)
print("SLICE 2: Weightlifting Minutes (Last Two Friends)")
print("=" * 50)
for index, minutes in enumerate(weightlifting_last_two, start=len(workout_list)-2):
    print(f"Friend {index}: Weightlifting = {minutes} min")
print("=" * 50)
```

```python
# Check if any friend's total workout minutes are >= 120
ACTIVE_THRESHOLD = 120  # Constant for clarity and maintainability

print("\n" + "=" * 50)
print(f"ACTIVE MEMBERS CHECK (>= {ACTIVE_THRESHOLD} minutes)")
print("=" * 50)

active_count = 0
for friend_name, value in workout_stats.items():
    # Check if this is workout data (tuple) and not a total entry
    if isinstance(value, tuple):
        total_workout_minutes = sum(value)
        if total_workout_minutes >= ACTIVE_THRESHOLD:
            print(f"Great job staying active, {friend_name}!")
            active_count += 1

if active_count == 0:
    print("No friends have reached the 120-minute threshold yet.")
    
print("=" * 50)
```

```python
# User input feature: Search for a friend's workout statistics
print("\n" + "=" * 50)
print("FRIEND LOOKUP SYSTEM")
print("=" * 50)
print(f"Available friends: {', '.join([name for name in workout_stats.keys() if not name.endswith('_Total')])}")
print("-" * 50)

# Get user input
search_name = input("Enter a friend's name to view their workout stats: ").strip()

# Search and display results
if search_name in workout_stats and isinstance(workout_stats[search_name], tuple):
    workout_data = workout_stats[search_name]
    total_key = f"{search_name}_Total"
    
    print(f"\nWorkout Statistics for {search_name}:")
    print("-" * 50)
    print(f"  • Yoga:          {workout_data[0]} minutes")
    print(f"  • Running:       {workout_data[1]} minutes")
    print(f"  • Weightlifting: {workout_data[2]} minutes")
    print("-" * 50)
    print(f"  TOTAL:           {workout_stats[total_key]} minutes")
    print("=" * 50)
else:
    print(f"\nFriend '{search_name}' not found in the records.")
    print("=" * 50)
```

```python
# Filter dictionary to get only total entries
total_entries = {
    key: value 
    for key, value in workout_stats.items() 
    if key.endswith('_Total')
}

# Find friends with highest and lowest total workout minutes
if total_entries:
    max_total_key = max(total_entries, key=total_entries.get)
    min_total_key = min(total_entries, key=total_entries.get)
    
    # Extract friend names by removing "_Total" suffix
    most_active_friend = max_total_key.replace('_Total', '')
    least_active_friend = min_total_key.replace('_Total', '')
    
    # Display results
    print("\n" + "=" * 50)
    print("WORKOUT STATISTICS SUMMARY")
    print("=" * 50)
    print(f"Most Active Friend:  {most_active_friend}")
    print(f"Total Minutes:       {total_entries[max_total_key]} minutes")
    print("-" * 50)
    print(f"Least Active Friend: {least_active_friend}")
    print(f"Total Minutes:       {total_entries[min_total_key]} minutes")
    print("=" * 50)
else:
    print("\nNo workout data available for analysis.")
    print("=" * 50)
```
